# - Define satisfaction dummy (1 if satislife ∈ [1, 2]); aggregate by country-year and compute logit-transformed satisfaction share

````python
df_filtered = df_filtered.copy()

# create dummy variable for life satisfaction
df_filtered["satis_dummy"] = df_filtered["satislife"].isin([1, 2]).astype(int)

# aggregate from individual responses by country and by year
agg_df = df_filtered.groupby(["country", "year"]).agg(
    share_satislife=("satis_dummy", "mean"),
    count=("satis_dummy", "size")
).reset_index()

# transform life satisfaction => logit transformation
epsilon = 1e-6
agg_df["logit_satislife"] = np.log(
    (agg_df["share_satislife"] + epsilon) / (1 - agg_df["share_satislife"] + epsilon)
)

# create quarterly variable to match frequencies
agg_df["year"] = agg_df["year"].astype(int)
agg_df["quarter_year"] = agg_df["year"].astype(str)+ "Q2"

# quick check
agg_df.head()
````

````python
# Mapping von Country-IDs auf Ländernamen
country_map = {
    1: "France",
    2: "Belgium",
    3: "Netherlands",
    4: "Germany",  # West
    5: "Italy",
    6: "Luxembourg",
    8: "Ireland",
    11: "Greece",
    12: "Spain",
    13: "Portugal",
    14: "Germany",  # East
    16: "Finland",
    18: "Austria",
    19: "Cyprus",
    21: "Estonia",
    23: "Latvia",
    24: "Lithuania",
    25: "Malta",
    27: "Slovakia",
    28: "Slovenia",
}

# Erst Länder-ID in Namen umwandeln
agg_df["country"] = agg_df["country"].map(country_map)

# Gruppieren nach Ländernamen und beschreibende Statistik berechnen
desc_by_country = agg_df.groupby("country")["logit_satislife"].describe()

# Alphabetisch sortieren
desc_by_country = desc_by_country.sort_index()

# Anzeige aller Zeilen
pd.set_option('display.max_rows', None)

# Ausgabe
print(desc_by_country)
````

# 3. VARIABLE TRANSFORMATION
#   - Create lagged variables: loan demand, credit standards, macro variables

````python
# create lagged variables
df_lagged = df_model.copy()
df_lagged["loan_demand_pct_lag1"] = df_lagged.groupby("country")["loan_demand_pct"].shift(1)
df_lagged["credit_standards_pct_lag1"] = df_lagged.groupby("country")["credit_standards_pct"].shift(1)

# baseline lagged model
df_baseline_lagged = df_lagged.dropna(subset=[
    "logit_satislife", "loan_demand_pct_lag1", "credit_standards_pct_lag1"
]).copy()

X = df_baseline_lagged[["loan_demand_pct_lag1", "credit_standards_pct_lag1"]]
X = sm.add_constant(X)
y = df_baseline_lagged["logit_satislife"]

# OLS output
model_baseline_lagged = sm.OLS(y, X).fit()
print("=== Regressionszusammenfassung ===")
print(model_baseline_lagged.summary())

# save results
with open("Lagged_Model_Summary.txt", "w") as f:
    f.write(model_baseline_lagged.summary().as_text())

# predictions
df_baseline_lagged["yhat_lagged"] = model_baseline_lagged.predict(X)

# plot loan demand
plt.figure(figsize=(8, 6))
sns.regplot(
    x="loan_demand_pct_lag1",
    y="yhat_lagged",
    data=df_baseline_lagged,
    scatter_kws={"alpha": 0.6, "color": "gray"},
    line_kws={"color": "green", "linewidth": 2}
)
plt.title("Loan Demand pct (t−1) vs Predicted Life Satisfaction", fontsize=14)
plt.xlabel("Loan Demand pct (t−1)", fontsize=12)
plt.ylabel("Predicted logit(Life Satisfaction)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("LoanDemand_Lagged_Prediction.png", dpi=300)
plt.show()  

# plot credit standards
plt.figure(figsize=(8, 6))
sns.regplot(
    x="credit_standards_pct_lag1",
    y="yhat_lagged",
    data=df_baseline_lagged,
    scatter_kws={"alpha": 0.6, "color": "gray"},
    line_kws={"color": "red", "linewidth": 2}
)
plt.title("Credit Standards pct (t−1) vs Predicted Life Satisfaction", fontsize=14)
plt.xlabel("Credit Standards pct (t−1)", fontsize=12)
plt.ylabel("Predicted logit(Life Satisfaction)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("CreditStandards_Lagged_Prediction.png", dpi=300)
plt.show()  
````

# - Fixed Effects Model 

````python
# create lagged variables
df_model_lagged = df_model.copy()
df_model_lagged["loan_demand_pct_lag1"] = df_model_lagged.groupby("country")["loan_demand_pct"].shift(1)
df_model_lagged["credit_standards_pct_lag1"] = df_model_lagged.groupby("country")["credit_standards_pct"].shift(1)

# dop NAs
df_model_lagged_clean = df_model_lagged.dropna(subset=[
    "logit_satislife",
    "loan_demand_pct_lag1",
    "credit_standards_pct_lag1"
]).copy()

# fixed effect regression
model_fe_lagged = smf.ols(
    "logit_satislife ~ loan_demand_pct_lag1 + credit_standards_pct_lag1 + C(country) + C(quarter_year)",
    data=df_model_lagged_clean
)
results_fe_lagged = model_fe_lagged.fit()

# OLS output
print(results_fe_lagged.summary())

# save results
with open("OLS_FE_Lagged_Summary.txt", "w") as f:
    f.write(results_fe_lagged.summary().as_text())

# predictions
df_model_lagged_clean["yhat_fe_lagged"] = results_fe_lagged.fittedvalues
````

````python
#CS Plot

# estimate model again
model_fe_lagged = smf.ols(
    "logit_satislife ~ loan_demand_pct_lag1 + credit_standards_pct_lag1 + C(country) + C(quarter_year)",
    data=df_model_lagged_clean
).fit()

# Residual plot (Partial Plot Approach)
# partial regression
resid_y = smf.ols(
    "logit_satislife ~ loan_demand_pct_lag1 + C(country) + C(quarter_year)",
    data=df_model_lagged_clean
).fit().resid

# now regress credit_standards_pct_lag1 on rest
resid_x = smf.ols(
    "credit_standards_pct_lag1 ~ loan_demand_pct_lag1 + C(country) + C(quarter_year)",
    data=df_model_lagged_clean
).fit().resid

# plot residuals (shows partial correlation)
plt.figure(figsize=(8, 6))
sns.regplot(
    x=resid_x,
    y=resid_y,
    scatter_kws={"alpha": 0.6, "color": "gray"},
    line_kws={"color": "red", "linewidth": 2}
)
plt.title("Partial Effect of Credit Standards (t−1) on Life Satisfaction (with Fixed Effects)")
plt.xlabel("Residuals of Credit Standards pct (t−1)")
plt.ylabel("Residuals of Logit Life Satisfaction")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

# 5 BASELINE MODEL (loan demand and credit standards lags + country and time fixed effects+standardized)

````python
###baseline model###

from sklearn.preprocessing import StandardScaler

# create lagged variables
df_model_lagged = df_model.copy()
df_model_lagged["loan_demand_pct_lag1"] = df_model_lagged.groupby("country")["loan_demand_pct"].shift(1)
df_model_lagged["credit_standards_pct_lag1"] = df_model_lagged.groupby("country")["credit_standards_pct"].shift(1)

# remove NAs
df_model_lagged_clean = df_model_lagged.dropna(subset=[
    "logit_satislife", "loan_demand_pct_lag1", "credit_standards_pct_lag1"
]).copy()

# standardize variables
scaler = StandardScaler()
df_model_lagged_clean["loan_demand_lag_std"] = scaler.fit_transform(df_model_lagged_clean[["loan_demand_pct_lag1"]])
df_model_lagged_clean["credit_standards_lag_std"] = scaler.fit_transform(df_model_lagged_clean[["credit_standards_pct_lag1"]])

# fixed effects OLS 
model_fe_std = smf.ols(
    "logit_satislife ~ loan_demand_lag_std + credit_standards_lag_std + C(country) + C(quarter_year)",
    data=df_model_lagged_clean
).fit()

# output
print(model_fe_std.summary())

# save results
with open("Final_Model3_FE_Lagged_Standardized.txt", "w") as f:
    f.write(model_fe_std.summary().as_text())

# predictions
df_model_lagged_clean["yhat_model3"] = model_fe_std.fittedvalues
````

# Robustness checks baseline model


````python
# robustness checks
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.stattools import jarque_bera
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.api as sm

# plotting residuals vs fitted values to assess heterogenity-homgenity baseline model

residuals = model_fe_std.resid
fitted_values = model_fe_std.fittedvalues
residuals = model_fe_std.resid

plt.figure(figsize=(8, 6))
plt.scatter(fitted_values, residuals, alpha=0.6)
plt.axhline(0, color='red', linestyle='--', linewidth=2)
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Breusch Pagan test for heteroscedasticity
_, pval, _, _ = het_breuschpagan(model_fe_std.resid, model_fe_std.model.exog)
print(f"Breusch-Pagan p-value: {pval:.4f}")

# Durbin-Watson test for autocorrelation in residuals
residuals = model_fe_std.resid

dw_stat = durbin_watson(residuals)
print(f"Durbin-Watson Teststatistik: {dw_stat:.4f}")

# plot lagged residuals vs residuals
residuals = model_fe_std.resid
residuals_lagged = pd.Series(residuals).shift(1)

plt.figure(figsize=(8, 6))
plt.scatter(residuals_lagged[1:], residuals[1:], alpha=0.6)
plt.title('Residuals vs. Lagged Residuals (Autocorrelation Check)')
plt.xlabel('Lagged Residuals (t-1)')
plt.ylabel('Residuals (t)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.axvline(0, color='red', linestyle='--')
plt.show()

# histogram of residuals
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# QQ plot 
residuals = model_fe_std.resid

plt.figure(figsize=(8, 6))
sm.qqplot(residuals, line='45', fit=True)
plt.title('Q-Q Plot of Residuals')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Jarque Bera test
residuals = model_fe_std.resid
jb_test = jarque_bera(residuals)

print(f"Jarque-Bera Teststatistik: {jb_test[0]:.4f}")
print(f"p-Wert: {jb_test[1]:.4f}")
print(f"Schiefe: {jb_test[2]:.4f}")
print(f"Kurtosis: {jb_test[3]:.4f}")

# check multicollinearity
X = model_fe_std.model.exog
vif = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
print(vif)
````

````python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.stattools import durbin_watson, jarque_bera
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
import seaborn as sns

def diagnose_model(model, model_name=""):
    print(f"\n{'='*30}\nDiagnostics for {model_name}\n{'='*30}")

    # Robust standard errors (HC3)
    robust_model = model.get_robustcov_results(cov_type='HC3')
    print(f"\nRobust Standard Errors (HC3):")
    print(robust_model.summary())

    # Robust standard errors (Newey-West)
    robust_model_hac = model.get_robustcov_results(cov_type='HAC', maxlags=1)
    print(f"\nRobust Standard Errors (HAC, Newey-West):")
    print(robust_model_hac.summary())

    residuals = model.resid
    exog = model.model.exog

    # Breusch-Pagan test
    try:
        exog_bp = sm.add_constant(exog, has_constant='add')  # required for BP test
        _, pval, _, _ = het_breuschpagan(residuals, exog_bp)
        print(f"\nBreusch-Pagan p-value: {pval:.4f} {'(Heteroskedasticity!)' if pval <= 0.05 else '(OK)'}")
    except Exception as e:
        print(f"\nBreusch-Pagan test not executable: {e}")

    # Durbin-Watson test
    dw_stat = durbin_watson(residuals)
    print(f"Durbin-Watson: {dw_stat:.4f} → {'positive autocorrelation!' if dw_stat < 1.5 else ('negative autocorrelation!' if dw_stat > 2.5 else 'OK')}")

    # Jarque-Bera test
    jb_test = jarque_bera(residuals)
    print(f"\nJarque-Bera Test:")
    print(f"  Test statistic: {jb_test[0]:.4f}")
    print(f"  p-value: {jb_test[1]:.4f} → {'normally distributed' if jb_test[1] > 0.05 else 'not normally distributed'}")
    print(f"  Skewness: {jb_test[2]:.4f}")
    print(f"  Kurtosis: {jb_test[3]:.4f}")

    # VIF (only if multiple predictors)
    exog_names = model.model.exog_names
    real_predictors = [name for name in exog_names if not name.startswith('C(') and name != 'Intercept']
    if len(real_predictors) > 1:
        cols_idx = [exog_names.index(name) for name in real_predictors]
        X_real = exog[:, cols_idx]
        vif_values = [variance_inflation_factor(X_real, i) for i in range(X_real.shape[1])]
        vif_df = pd.DataFrame({'Variable': real_predictors, 'VIF': vif_values})
        print("\nVariance Inflation Factors (VIF):")
        print(vif_df)
    else:
        print("\nVIF: Only one predictor – multicollinearity cannot be assessed.")

    # Residual plots
    fitted_values = model.fittedvalues
    residuals_series = pd.Series(residuals)

    # Residuals vs. Fitted Values
    plt.figure(figsize=(8, 6))
    plt.scatter(fitted_values, residuals, alpha=0.6)
    plt.axhline(0, color='red', linestyle='--', linewidth=2)
    plt.title(f'Residuals vs. Fitted Values ({model_name})', fontdict={'weight': 'bold'})
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # Residuals vs. Lagged Residuals
    residuals_lagged = residuals_series.shift(1)
    plt.figure(figsize=(8, 6))
    plt.scatter(residuals_lagged[1:], residuals_series[1:], alpha=0.6)
    plt.title(f'Residuals vs. Lagged Residuals ({model_name})', fontdict={'weight': 'bold'})
    plt.xlabel('Residuals (t-1)')
    plt.ylabel('Residuals (t)')
    plt.axhline(0, color='red', linestyle='--')
    plt.axvline(0, color='red', linestyle='--')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # Histogram of Residuals
    plt.figure(figsize=(8, 6))
    sns.histplot(residuals, kde=True)
    plt.title(f'Histogram of Residuals ({model_name})', fontdict={'weight': 'bold'})
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # QQ Plot
    plt.figure(figsize=(8, 6))
    sm.qqplot(residuals, line='45', fit=True)
    plt.title(f'QQ Plot of Residuals ({model_name})', fontdict={'weight': 'bold'})
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Example usage:
diagnose_model(model_macro_no_intercept, model_name="Macro Model")
diagnose_model(model_baseline_no_intercept, model_name="Baseline Model")
````
